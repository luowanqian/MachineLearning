# Machine Learning
## Regression
### Linear Regression

__Import python package__

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import linear_model
from sklearn.metrics import r2_score
%matplotlib inline
```

__Load data and plot data__

```python
df = pd.read_csv('linear_regression_df.csv')
X = np.asarray(df.X).reshape(-1, 1)
Y = np.asarray(df.Y)
plt.xlabel('Explanatory')
plt.ylabel('Response')
plt.plot(X, Y, 'ro')
```

__Train data and make prediction__

```python
trainX = X[20:, ]
trainY = Y[20:]
testX = X[:20]
testY = Y[:20]
linear = linear_model.LinearRegression()
linear.fit(trainX, trainY)
prediction = linear.predict(testX)
score = r2_score(testY, prediction)
print('Coefficient: {}'.format(linear.coef_))
print('Intercept: {}'.format(linear.intercept_))
print('R2 socre: {}'.format(score))
```

__Plot predicted data__

```python
plt.plot(X, Y, 'ro')
plt.plot(testX, prediction, 'b')
plt.xlabel('Explanatory')
plt.ylabel('Response')
```
